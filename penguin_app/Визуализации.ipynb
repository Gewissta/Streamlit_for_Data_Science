```python
# импортируем pandas и seaborn
import pandas as pd
import seaborn as sns

# на экранах с высоким разрешением типа Retina 
# графики в тетрадках по умолчанию выглядят 
# размытыми, используем эту настройку 
# для улучшения резкости
%config InlineBackend.figure_format = 'retina'
```

```python
# загрузим данные
penguins_df = pd.read_csv('penguins.csv')
# посмотрим имена переменных
print(list(penguins_df.columns))
```

```python
# строим диаграмму рассеяния
sns.scatterplot(data=penguins_df, 
                x="длина_ласт_мм", 
                y="масса_тела_г",
                hue="вид");
```
