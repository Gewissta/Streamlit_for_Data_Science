```python
import requests
import json
from datetime import datetime, timedelta
import numpy as np
import pandas as pd
import re
```

```python
current_time = datetime.now()
previous_day = current_time - timedelta(days=5)
previous_day = datetime(previous_day.year, previous_day.month, previous_day.day, 19, 0)
start_time = previous_day.strftime("%Y-%m-%dT%H")
```

```python
response = requests.get(f"https://inkalika.99ns.ru/api/v1/treatments.json?find[created_at][$gte]={start_time}")
ins_lst = []
if response.status_code == 200:
    treatments_data = response.json()
    for i, x in enumerate(treatments_data):
        if x['eventType'] in ['Meal Bolus', 'Correction Bolus', 'Temp Basal']:
            time_obj = datetime.strptime(x['created_at'], "%Y-%m-%dT%H:%M:%S.%f%z")
            if 'rate' in x:
                ins_lst.append([time_obj, x['eventType'], x['insulin'], x['rate']])
            else:
                ins_lst.append([time_obj, x['eventType'], x['insulin'], None])

    df_ins = pd.DataFrame(ins_lst, columns=['Time', 'eventType', 'InsulinInjection', 'TBR'])
    df_ins.loc[df_ins['InsulinInjection'].isna(), 'InsulinInjection'] = 0
df_ins
```
